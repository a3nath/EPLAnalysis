```python

##importing libraries
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup, Comment
import re
import sqlite3 as db
conn = db.connect('../src/data/my_database.db')
pd.set_option('display.max_columns', 40)
```

```python
url_names = ["passing", "passing_types", "shooting", "possession", "defense", "gca"]
for name in url_names:
    URL = f'https://fbref.com/en/comps/9/{name}/Premier-League-Stats'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="all_stats_passing")
    comments = soup.findAll(text=lambda text:isinstance(text, Comment))
    df = pd.read_html(comments[30], header=[0,1])[0]
    df.columns = [col[1] if "Unnamed" in col[0] else '_'.join(col) for col in df.columns.values]
    df_clean = df.drop(columns = ["RK", "Nation", "Pos", "Age", "Born", "Matches"], errors="ignore")
    df_clean = df_clean[~df_clean["Player"].str.contains("Player")]
    columns = df_clean.iloc[:,4:].columns
    for col in columns:
        df_clean[col] = df_clean[col].astype(float)
    df_clean.to_sql(name, conn, if_exists="replace")
```
